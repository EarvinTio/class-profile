```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
salary = pd.read_csv('salary.csv', header=None)
```

```python
salary
```

```python
for term in range(6):
    print(len(salary[salary[2*term+7]=='cad'][term]))
    print(len(salary[salary[2*term+7]=='usd'][term]))
    print(len(salary[salary[2*term+7].isna()][term]))
    print(len(salary[salary[2*term+7]=='cad'][term]) + len(salary[salary[2*term+7]=='usd'][term]) + len(salary[salary[2*term+7].isna()][term]))
    print()
```

```python
term = 0 # check [0,1,2,3,4,5]
salary[salary[2*term+7]=='cad'][term] # check ='cad' , ='usd' , and .isna()
```

```python
cad, usd = [], []

for term in range(6):
    cad.append(salary[salary[2*term+7]=='cad'][2*term+6])
    usd.append(salary[salary[2*term+7]=='usd'][2*term+6])
```

```python
means_cad, n_cad = [], []
for term in cad:
    print(len(term))
    print(term.mean())
    n_cad.extend([len(term)])
    means_cad.extend([term.mean()])
print(n_cad)
print(means_cad)
```

```python
means_usd, n_usd = [], []
for term in usd:
    print(len(term))
    print(term.mean())
    n_usd.extend([len(term)])
    means_usd.extend([term.mean()])
print(n_usd)
print(means_usd)
```

```python
salary_cad = cad[0].tolist()
salary_cad.extend(cad[1].tolist())
salary_cad.extend(cad[2].tolist())
salary_cad.extend(cad[3].tolist())
salary_cad.extend(cad[4].tolist())
salary_cad.extend(cad[5].tolist())
```

```python
terms_cad = [1 for i in range(len(cad[0]))]
terms_cad.extend([2 for i in range(len(cad[1]))])
terms_cad.extend([3 for i in range(len(cad[2]))])
terms_cad.extend([4 for i in range(len(cad[3]))])
terms_cad.extend([5 for i in range(len(cad[4]))])
terms_cad.extend([6 for i in range(len(cad[5]))])
```

```python
new_cad = pd.DataFrame(list(zip(terms_cad,salary_cad)), columns=["terms","salary"])
```

```python
new_cad
```

```python
salary_usd = usd[0].tolist()
salary_usd.extend(usd[1].tolist())
salary_usd.extend(usd[2].tolist())
salary_usd.extend(usd[3].tolist())
salary_usd.extend(usd[4].tolist())
salary_usd.extend(usd[5].tolist())
```

```python
terms_usd = [1 for i in range(len(usd[0]))]
terms_usd.extend([2 for i in range(len(usd[1]))])
terms_usd.extend([3 for i in range(len(usd[2]))])
terms_usd.extend([4 for i in range(len(usd[3]))])
terms_usd.extend([5 for i in range(len(usd[4]))])
terms_usd.extend([6 for i in range(len(usd[5]))])
```

```python
new_usd = pd.DataFrame(list(zip(terms_usd,salary_usd)), columns=["terms","salary"])
```

```python
new_usd
```

```python
ax = sns.lineplot('terms','salary', data=new_cad)
x = [1,2,3,4,5,6]

ax.scatter(x,means_cad)

for i, txt in enumerate(n_cad):
    ax.annotate('n='+str(txt), (x[i], means_cad[i]), xytext=(x[i]-0.15,means_cad[i]-1.3))

for i, txt in enumerate(['%.2f'% mean for mean in means_cad]):
    ax.annotate('$'+str(txt), (x[i], means_cad[i]), xytext=(x[i]-0.6,means_cad[i]+1))

ax.set_xlim([0,7])
ax.set_xticklabels(['','1A','1B','2A','2B','3A','3B',''])
ax.set_ylim([15,31])
```

```python
ax = sns.lineplot('terms','salary', data=new_usd)

ax.scatter(x,means_usd)

for i, txt in enumerate(n_usd):
    ax.annotate('n='+str(txt), (x[i], means_usd[i]), xytext=(x[i]-0.1,means_usd[i]-3))

for i, txt in enumerate(['%.2f'% mean for mean in means_usd]):
    ax.annotate('$'+str(txt), (x[i], means_usd[i]), xytext=(x[i]-0.6,means_usd[i]+2.5))
    
ax.set_xlim([0,7])
ax.set_xticklabels(['','1A','1B','2A','2B','3A','3B',''])
ax.set_ylim([10,55])
```
