```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set(style="white")
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

data = pd.read_csv('responses.csv')
data.head()

labels = ["1A","1B","2A","2B","3A","3B","4A"]
focus = {"shade": True, "linewidth": 0, "alpha": 1, "zorder": 3}
fig, axs = plt.subplots(4,2,figsize=(15,20))
for i,ax in enumerate(axs.flatten()):        
    ax.set_yticks([])
    
    if i==0:
        sns.distplot( data.avg_1a, hist=False, kde_kws={"shade": True}, label=labels[0], ax=ax)
        sns.distplot( data.avg_1b, hist=False, kde_kws={"shade": True}, label=labels[1], ax=ax)
        sns.distplot( data.avg_2a, hist=False, kde_kws={"shade": True}, label=labels[2], ax=ax)
        sns.distplot( data.avg_2b, hist=False, kde_kws={"shade": True}, label=labels[3], ax=ax)
        sns.distplot( data.avg_3a, hist=False, kde_kws={"shade": True}, label=labels[4], ax=ax)
        sns.distplot( data.avg_3b, hist=False, kde_kws={"shade": True}, label=labels[5], ax=ax)
        sns.distplot( data.avg_4a, hist=False, kde_kws={"shade": True}, label=labels[6], ax=ax)
        ax.set_xlabel('All Terms')
    else:
        if i==1:
            sns.distplot( data.avg_1a, hist=False, kde_kws=focus, ax=ax)
            ax.axvline(data.avg_1a.median(), ls='--', color='black',zorder=11)
            ax.text(data.avg_1a.median()+1, 0.079, "median = "+str(data.avg_1a.median()))
        else:
            sns.distplot( data.avg_1a, hist=False, kde_kws={"shade": True, "alpha": 0.1}, ax=ax)
        if i==2:
            sns.distplot( data.avg_1b, hist=False, kde_kws=focus, ax=ax)
            ax.axvline(data.avg_1b.median(), ls='--', color='black',zorder=11)
            ax.text(data.avg_1b.median()+1, 0.079, "median = "+str(data.avg_1b.median()))
        else:
            sns.distplot( data.avg_1b, hist=False, kde_kws={"shade": True, "alpha": 0.1}, ax=ax)
        if i==3:
            sns.distplot( data.avg_2a, hist=False, kde_kws=focus, ax=ax)
            ax.axvline(data.avg_2a.median(), ls='--', color='black',zorder=11)
            ax.text(data.avg_2a.median()+1, 0.079, "median = "+str(data.avg_2a.median()))
        else:
            sns.distplot( data.avg_2a, hist=False, kde_kws={"shade": True, "alpha": 0.1}, ax=ax)
        if i==4:
            sns.distplot( data.avg_2b, hist=False, kde_kws=focus, ax=ax)
            ax.axvline(data.avg_2b.median(), ls='--', color='black',zorder=11)
            ax.text(data.avg_2b.median()+1, 0.079, "median = "+str(data.avg_2b.median()))
        else:
            sns.distplot( data.avg_2b, hist=False, kde_kws={"shade": True, "alpha": 0.1}, ax=ax)
        if i==5:
            sns.distplot( data.avg_3a, hist=False, kde_kws=focus, ax=ax)
            ax.axvline(data.avg_3a.median(), ls='--', color='black',zorder=11)
            ax.text(data.avg_3a.median()+1, 0.079, "median = "+str(data.avg_3a.median()))
        else:
            sns.distplot( data.avg_3a, hist=False, kde_kws={"shade": True, "alpha": 0.1}, ax=ax)
        if i==6:
            sns.distplot( data.avg_3b, hist=False, kde_kws=focus, ax=ax)
            ax.axvline(data.avg_3b.median(), ls='--', color='black',zorder=11)
            ax.text(data.avg_3b.median()+1, 0.079, "median = "+str(data.avg_3b.median()))
        else:
            sns.distplot( data.avg_3b, hist=False, kde_kws={"shade": True, "alpha": 0.1}, ax=ax)
        if i==7:
            sns.distplot( data.avg_4a, hist=False, kde_kws=focus, ax=ax)
            ax.axvline(data.avg_4a.median(), ls='--', color='black',zorder=11)
            ax.text(data.avg_4a.median()+1, 0.079, "median = "+str(data.avg_4a.median()))
        else:
            sns.distplot( data.avg_4a, hist=False, kde_kws={"shade": True, "alpha": 0.1}, ax=ax)
        ax.set_xlabel(labels[i-1]+' Average')
    ax.axhline(y=-.000001, color='black', lw=2, zorder=10)
    sns.despine(bottom=True,left=True)

plt.savefig('term_averages.png')
```
