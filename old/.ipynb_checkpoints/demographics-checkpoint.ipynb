```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
data = pd.read_csv('responses.csv')
```

```python
def plotPie(category):
    # Pie chart
    if category == 'politics':
        labels = ['Liberal', 'Neutral', 'Extremely Liberal', 'Conservative']
    else:
        labels = data[category].value_counts().index.tolist()
    sizes = np.array(data[category].value_counts().values.tolist())/64*100
    sizes.tolist()
    
    #colors
    # colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

    fig1, ax1 = plt.subplots(figsize=(10,7))
    pie = ax1.pie(sizes, radius=1, autopct='%1.1f%%',rotatelabels =True, counterclock=False, startangle=90)

    #draw circle
    centre_circle = plt.Circle((0,0),0.70,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)

    # [0] = wedges, [1] = labels, [2] = fractions
    # Set font size
    labels_text_list = pie[1]
    for text in labels_text_list:
        text.set_fontsize(12)
    
    # Rotate fractions
    fraction_text_list = pie[2]
    for text in fraction_text_list:
        text.set_rotation(315)
        text.set_fontsize(12)

    # Equal aspect ratio ensures that pie is drawn as a circle
    ax1.axis('equal')
    plt.legend(pie[0], labels, loc="best")
    plt.tight_layout()
```

```python
plotPie('birth')
plt.savefig('birth.png')
```

```python
plotPie('gender')
plt.savefig('gender.png')
```
