```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from collections import Counter
from scipy.stats import ttest_ind
from statannot import add_stat_annotation

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

data = pd.read_csv('responses.csv')
data.head()

print(list(set(data['ethnicity'].tolist())))
print()
print(Counter(data['ethnicity'].tolist()))
print()

seasian = data.loc[data['ethnicity'] == "Southeast Asian"]
easian = data.loc[data['ethnicity'] == "East Asian"]
caucasian = data.loc[data['ethnicity'] == "Cauasian"]
meast = data.loc[data['ethnicity'] == "Middle Eastern"]
sasian = data.loc[data['ethnicity'] == "South Asian"]
mixed = data.loc[data['ethnicity'] == "Mixed"]

seasian_grades = seasian[['avg_1a', 'avg_1b', 'avg_2a', 'avg_2b', 'avg_3a', 'avg_3b', 'avg_4a']].mean(axis=1)
seasian_labels = pd.Series(["Southeast Asian" for i in range(len(seasian_grades))])

easian_grades = easian[['avg_1a', 'avg_1b', 'avg_2a', 'avg_2b', 'avg_3a', 'avg_3b', 'avg_4a']].mean(axis=1)
easian_labels = pd.Series(["East Asian" for i in range(len(easian_grades))])

caucasian_grades = caucasian[['avg_1a', 'avg_1b', 'avg_2a', 'avg_2b', 'avg_3a', 'avg_3b', 'avg_4a']].mean(axis=1)
caucasian_labels = pd.Series(["Caucasian" for i in range(len(caucasian_grades))])

meast_grades = meast[['avg_1a', 'avg_1b', 'avg_2a', 'avg_2b', 'avg_3a', 'avg_3b', 'avg_4a']].mean(axis=1)
meast_labels = pd.Series(["Middle Eastern" for i in range(len(meast_grades))])

sasian_grades = sasian[['avg_1a', 'avg_1b', 'avg_2a', 'avg_2b', 'avg_3a', 'avg_3b', 'avg_4a']].mean(axis=1)
sasian_labels = pd.Series(["South Asian" for i in range(len(sasian_grades))])

mixed_grades = mixed[['avg_1a', 'avg_1b', 'avg_2a', 'avg_2b', 'avg_3a', 'avg_3b', 'avg_4a']].mean(axis=1)
mixed_labels = pd.Series(["Mixed" for i in range(len(mixed_grades))])

grades = seasian_grades.append(
         easian_grades,ignore_index=True).append(
         caucasian_grades,ignore_index=True).append(
         meast_grades,ignore_index=True).append(
         sasian_grades,ignore_index=True).append(
         mixed_grades,ignore_index=True)
labels = seasian_labels.append(
         easian_labels,ignore_index=True).append(
         caucasian_labels,ignore_index=True).append(
         meast_labels,ignore_index=True).append(
         sasian_labels,ignore_index=True).append(
         mixed_labels,ignore_index=True)

x = "Ethnicity"
y = "Average Grade"

df = pd.DataFrame({
    x: labels,
    y: grades
})

labels_sorted = sorted(list(set(df[x])))
order = labels_sorted

fig, ax = plt.subplots(figsize=(10,7))
p = sns.boxplot(x=x,y=y,order=order,data=df,showfliers=False)
p = sns.swarmplot(x=x,y=y,order=order,data=df,color=".2")
p.set_xticklabels(p.get_xticklabels(), rotation=45, horizontalalignment='right')

max_sorted = []
for i in order:
    data = df[df[x] == i][y]
    q1, q3 = data.quantile([0.25,0.75])
    max_sorted.append(max(data.where(data <= q3 + 1.5*(q3-q1))))

ns = [df[x].tolist().count(i) for i in order]

for i,xtick in enumerate(p.get_xticks()):
    p.text(xtick,max_sorted[i]+0.7,'n='+str(ns[i]),horizontalalignment='center',color='k',style='italic')

plt.tight_layout()
plt.savefig('ethnicity.png')
```
